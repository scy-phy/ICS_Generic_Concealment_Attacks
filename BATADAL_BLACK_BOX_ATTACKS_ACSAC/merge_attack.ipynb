```python
import pandas as pd 
data = pd.read_csv('test_dataset_1.csv')
data = data.drop(columns=['Unnamed: 0'], axis =1)
```

```python
attacks = data.loc[data['ATT_FLAG'] == 1]
```

```python
attacks
```

```python
indexes = attacks[1:].loc[(attacks.DATETIME[:-1].values-attacks.DATETIME[1:].values) < -0.25].index.insert(0, 6909).values
```

```python
attack_ids = [1,2,3,4,5,6,7]
data_mod = data.copy(deep=True)
for i in [2,3,4,5,6,7,8,9,10,15,20,25,30,35,40]:
    folder = "AE_max_concealable_var_"+str(i) #"unconstrained_attack"#
    file_name = "_max"+str(i) #''#
    for attack_id in attack_ids:
        print(attack_id)
        atck = pd.read_csv('./'+folder+'/new_advAE_attack_'+str(attack_id)+'_from_test_dataset'+file_name+'.csv')
        atck = atck.drop(columns=['Unnamed: 0'], axis=1)
        index = indexes[attack_id-1]
        print(index)
        if attack_id == 1:
            data_mod.loc[index:index+len(atck)-2, atck.columns] = atck.values[:-1]
        else:
            data_mod.loc[index+1:index+len(atck)-1, atck.columns] = atck.values[:-1]
        #print(data.loc[index]-data_mod.loc[index])

    data_mod.to_csv('./'+folder+'/test_dataset_1_unconstrained_newAE.csv', index=False)
```
